```python
import pandas as pd
import numpy as np
```

```python
full_dataset = pd.read_csv(r'/Users/allysonenglish/Desktop/workbench/working_data_7_days.csv', low_memory=False).drop(columns=['year', 'quarter', 'one_way', 'surface', 'road_type'])
```

```python
sample = full_dataset[:150]
```

```python
time_of_day = pd.Series([])
```

```python
for i in range(0, len(sample)):
    if sample['hour_of_day'][i] >= 22 or sample['hour_of_day'][i] <= 5:
        time_of_day[i] = "overnight"
    if sample['hour_of_day'][i] >= 6 and sample['hour_of_day'][i] <= 9:
        time_of_day[i] = "morning"
    if sample['hour_of_day'][i] >= 10 and sample['hour_of_day'][i] <= 16:
        time_of_day[i] = "midday"
    if sample['hour_of_day'][i] >= 17 and sample['hour_of_day'][i] <= 21:
        time_of_day[i] = "evening"
```

```python
sample.insert(0, "time_of_day", time_of_day)
```

```python
sample.set_index(['time_of_day', 'hour_of_day'])
```

```python
sample.groupby('time_of_day').mean()
```
